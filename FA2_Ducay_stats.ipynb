```python
## 3.90 Find the geometric mean of the sets (a) 3, 5, 8, 3, 7, 2 and (b) 28.5, 73.6, 47.2, 31.5, 64.8.

product = 1
a = [3,5,8,3,7,2]
for i in range(len(a)):
    product = product * a[i]
    n = i
n += 1
geom_a = (product)**(1/n)

product = 1
b = [28.5, 73.6, 47.2, 31.5, 64.8]
for i in range(len(b)):
    product = product * b[i]
    n = i
n += 1
geom_b = (product)**(1/n)

print(f"Geometric mean a: {geom_a}")
print(f"Geometric mean b: {geom_b}")
```
